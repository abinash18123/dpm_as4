```python

#fixed deposit
def fixed(amount,rates,s,e):
  rates = 1 + rates/100
  fd_df = pd.DataFrame()
  arr = [(str(i) + "-01-01")  for i in range(s,e+1)]
  arr = np.array(arr)
  fd_df['Date'] = arr.tolist()
  out = [0 for i in range(e-s+1)]
  for i in range(e-s+1):
    out[i] += amount 
    out = [i*rates for i in out]

    # print(out)
  out1 = out[::-1]
  for i in range(1,e-s+1):
    out1[i] += out1[i-1]
  out1 = np.array(out1)
  fd_df['FINAL_AMOUNT'] = out1.tolist()
  # print(fd_df)
  return fd_df

#for stock
def checker(stock,start,end,period,amount):
  stock = dic_t[stock]
  index = stock.index
  start_index = (stock[stock['Date']==start].index.values)
  end_index = (stock[stock['Date']==end].index.values)
  global numbers
  # print(start_index,end_index )
  if len(start_index) == 0 or len(end_index)==0:
    print("NOT VALID")
    return ("Not valid")
  else:
    start_index = start_index[0]
    end_index = end_index[0]
    range_weekly=  [i for i in range(start_index,end_index,5)]
    range_monthly=  [i for i in range(start_index,end_index,20)]
    range_quaterly=  [i for i in range(start_index,end_index,80)]
    if period == "weekly":
      duration = range_weekly
    elif period == "monthly":
      duration = range_monthly
    else:
      duration = range_quaterly
    weekly_data = stock.iloc[duration]
    # weekly_data


    weekly_data["CHANGE"] = weekly_data.Close.pct_change() +1

    weekly_data['CHANGE'] = weekly_data['CHANGE'].fillna(1)
    weekly_data["AMOUNT"] = int(amount) * weekly_data["CHANGE"]

  weekly_data.insert(0, 'indexx', range(0, 0 + len(weekly_data)))
  weekly_data.set_index('indexx')
  # print(weekly_data)
  coun = 0
# flag = int(amount)
  for index, row in weekly_data.iterrows():
      coun +=1
      flag = row['AMOUNT']
      if coun ==2:
        break
  arr = [int(amount)]
  coun = 0
  for index, row in weekly_data.iterrows():
      coun +=1
      temp = row['CHANGE']
      #  print("coun",coun)
      #  flag = row['AMOUNT']
      #  tempp = row['AMOUNT']
      if coun >2:
        check = ((flag+int(amount))) 
        #  print (check)
        flag = check*temp
        arr.append(check)
        #  print(flag,temp)
  arr.append(arr[-1] + 1000)
  arr = np.array(arr)

  arr
  numbers = len(arr)
  # print(len(arr))
  weekly_data['FINAL_AMOUNT'] = arr.tolist()


  return weekly_data

```

```python
# user input
print("PICK A STOCK")
stock = input()
print("Pick start date:")
start = input()
print("Pick end date:")
end = input()
print("Pick period(weekly/monthly/quaterly):")
period = input()
print("Pick an amount:")
amount = input()

output= checker(stock,start,end,period,amount)


output_file("dark_minimal.html")  #template

curdoc().theme = 'dark_minimal'


p = figure(title='dark_minimal',plot_width = 1400,plot_height  = 500,x_axis_type = "datetime")  #initialize plot
p.title.text = "CHANGE IN AMOUNT" #title

for d,ind,c in zip([output],[str(stock)],Spectral4): #data , legend name and colour 
  df1 = pd.DataFrame(d)  #read data

  df1['Date'] = pd.to_datetime(df1['Date'])  #read data
  p.line(df1["Date"],df1["FINAL_AMOUNT"],line_width = 1.5,color = c,legend_label = ind) #plot 

p.legend.location = "top_left"  

p.legend
output_notebook()
show(p)
```
